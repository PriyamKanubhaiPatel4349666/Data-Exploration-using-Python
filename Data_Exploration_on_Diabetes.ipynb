```python
# Load the diabetes dataset
df = pd.read_csv('Diabetes.csv')
```

```python
# Display the first few rows of the dataset
print("Original dataset:")
print(df.head())
```

# Data Cleaning


```python
# Check for missing values
print("\nMissing values before cleaning:")
print(df.isnull().sum())
```

```python
# Remove rows with missing values
data = df.dropna()
```

```python
# Display the first few rows of the cleaned dataset
print("\nCleaned dataset:")
print(data.head())
```

```python
# @ Check for missing values after cleaning
print("\nMissing values after cleaning:")
print(data.isnull().sum())
```

#Visualisations

# **Visualizing Age Distribution**
This code generates a histogram to show the distribution of ages in the dataset. The histogram, created using seaborn, has 20 bins and includes a kernel density estimate (KDE) curve for a smooth representation. The plot is colored sky blue and titled "Distribution of Age." The x-axis represents "Age," and the y-axis shows frequency. Finally, the plot is displayed using `plt.show()`.

```python
# Visualize the distribution of age
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

# **Diabetes Diagnosis Distribution**
This code visualizes diabetes diagnosis distribution using a count plot. Seaborn's `countplot` function creates the plot.

```python

# Visualize the distribution of diabetes diagnosis
plt.figure(figsize=(6, 6))
sns.countplot(x='class', data=data, palette='Set2')
plt.title('Distribution of Diabetes Diagnosis')
plt.xlabel('Diabetes Diagnosis')
plt.ylabel('Count')
plt.show()
```

# **Age Distribution**
This code visualizes age distribution with a histogram. Seaborn's `histplot` function generates the plot.

```python
# Visualize the distribution of age
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# **Age vs Plasma Glucose**
This code plots the relationship between age and plasma glucose concentration using a scatterplot.

```python
# Visualize the relationship between age and plasma glucose concentration
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='PG Concentration', data=data)
plt.title('Scatterplot of Age vs Plasma Glucose Concentration')
plt.xlabel('Age')
plt.ylabel('Plasma Glucose Concentration')
plt.show()
```

# **Age Distribution by Diabetes**
Violin plot visualizes age distribution by diabetes diagnosis. Seaborn's `violinplot` creates the plot.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='class', y='Age', data=data, palette='pastel')
plt.title('Age Distribution by Diabetes Diagnosis')
plt.xlabel('Diabetes Diagnosis')
plt.ylabel('Age')
plt.show()
```

# **BMI Distribution by Diabetes**
Boxplot visualizes BMI distribution by diabetes diagnosis. Seaborn's `boxplot` creates the plot.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='BMI', data=data, palette='pastel')
plt.title('BMI Distribution by Diabetes Diagnosis')
plt.xlabel('Diabetes Diagnosis')
plt.ylabel('BMI')
plt.show()
```

# **Pregnancy Count by Diabetes**
Bar plot shows pregnancy count by diabetes diagnosis. Seaborn's `barplot` creates the plot.

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='class', y='Pregnancies', data=data, palette='pastel')
plt.title('Pregnancy Count by Diabetes Diagnosis')
plt.xlabel('Diabetes Diagnosis')
plt.ylabel('Mean Pregnancy Count')
plt.show()
```

# **Glucose Distribution by Diabetes**
Kernel Density Estimate (KDE) plot shows glucose distribution by diabetes diagnosis.

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=data, x='PG Concentration', hue='class', fill=True, palette='pastel', common_norm=False)
plt.title('Glucose Distribution by Diabetes Diagnosis')
plt.xlabel('Plasma Glucose Concentration')
plt.ylabel('Density')
plt.show()
```

# **Pairwise Relationships by Diabetes**
PairGrid plots pairwise relationships by diabetes diagnosis. It combines scatter plots, KDE plots, and regression plots.

```python
g = sns.PairGrid(data, vars=['Age', 'BMI', 'DP Function'], hue='class', palette='husl')
g.map_upper(sns.scatterplot, alpha=0.7)
g.map_diag(sns.kdeplot, fill=True)
g.map_lower(sns.regplot, scatter_kws={'alpha':0.5}, line_kws={'color':'orange'})
g.add_legend(title='Diabetes Diagnosis')
plt.suptitle('Pairwise Relationships by Diabetes Diagnosis', y=1.02)
plt.show()
```
